```python
#import numpy as np
import pandas as pd

df = pd.read_csv('movies_2.csv') # Gathering data
df.head(10)



```

```python
# check for MPAA Rating, Budget, Gross, Release Date, Genre, Runtime, Rating and Rating Count columns how many NaN we have

df.iloc[ : , 2 : 10].isna().sum()
```

```python
# Get rid of the Summary column


df_clean = df.drop(df.columns[[10]], axis=1)

# Remove spaces from the column names. 

df_clean.columns = [c.replace(' ', '_') for c in df_clean.columns]
df_clean

```

```python
df_clean.info()
```

```python
# Search for duplicate Titles, sort the list and then print the list.


df_clean[df_clean.duplicated(["Title"], keep=False)]
# df_dup.sort_values(["Title"])



```

```python
# Remove the duplicate Title "Jurassic Park 3" and keep in place within the df_drop 

df_clean.drop([98], inplace=True)
```

```python
# Lets check if the row is removed from the df_drop

df_clean.iloc[98]
```

```python
# Changing the data type of the Release date only. As Budget and Gross are already type Float

df_clean["Release_Date"] = df_clean["Release_Date"].astype("datetime64")
df_clean.info()
```

```python
# Dropping the last row with the NaN values. 

df_clean.drop(615, inplace=True)

```

```python
# Checking to see the DF after the row is dropped. 
df_clean
```

```python
# Fill missing values of Rating with the median value.
df_clean["Rating"].fillna(value=df_clean["Rating"].median(), inplace=True)
```

```python
df_clean
```

```python
# Fill missing values of Rating Count with the median value.
df_clean["Rating_Count"].fillna(value=df_clean["Rating_Count"].median(), inplace=True)
```

```python
df_clean
```

```python
# Show the movies with more than 7 in Rating & greater than 50 million Gross
df_clean.loc[(df_clean["Rating"] > 7) & (df_clean["Gross"] > 50000000)]
```

```python
# count of Animation movies with more than 7 in Rating (use the shape() function)

df_anime = df_clean.loc[(df_clean["Genre"] == "Animation") & (df_clean["Rating"] > 7)]
df_anime.shape
```
