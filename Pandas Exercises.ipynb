```python
# Show the movies with more than 7 in Rating & greater than 50 million Gross
df.loc[(df["Rating"] > 7) & (df["Gross"] > 50000000)]
```

```python
# show the movies with more than 7 in Rating & greater than 50 million Gross & with Parental guidance as MPAA Rating
df.loc[(df["Rating"] > 8) & (df["Gross"] > 50000000) & (df["MPAA Rating"] == "PG")]
```

```python
# count of Animation movies with more than 7 in Rating (use the shape() function)

df_anime = df.loc[(df["Genre"] == "Animation") & (df["Rating"] > 7)]
df_anime.shape
```

```python
# show the list of top 5 movies based on Budget

df_topBudget = df.sort_values("Budget", ascending = False)
df_topBudget.head()
```

```python
# show the top 5 Comedy movies by Rating


df_comedy = df.loc[(df["Genre"] == "Comedy")]
df_comedy.sort_values("Rating", ascending=False)
df_comedy.head()
```

```python
# top 5 movie names by Rating

df_top = df.sort_values("Rating", ascending=False)
df_top.head()
```

```python
# top 3 high Gross Romance movies released after 1999, not included (typecast it to datetime)


df_romance = df.loc[(df["Genre"] == "Romance")]
df_romance = df_romance.loc[(df_romance["Release Date"] > "2000-01-01")]
df_romance = df_romance.sort_values("Gross", ascending=False)
df_romance.head(3)
#df_romance
```

```python
# how many Genres are present in the dataframe? (use the function value_counts() which applies to Series, not Dataframe)
df.value_counts("Genre")
```

```python
# top 5 expensive movies released after 1999, not included (measured by Budget)

# df_romance = df.loc[(df["Genre"] == "Romance")]
# df_romance = df_romance.loc[(df_romance["Release Date"] > "2000-01-01")]

df_exp = df.loc[(df["Release Date"] > "2000-01-01")]
df_exp = df_exp.sort_values("Budget", ascending=False)
df_exp.head(5)
```

```python
# most & least frequent MPAA Rating in the dataset in terms of occurances

print(df['MPAA Rating'].mode()) # Most frequent using Mode

print(df['MPAA Rating'].value_counts().idxmax()) # Most frequent
print(df['MPAA Rating'].value_counts().idxmin()) # Least frequent
```
